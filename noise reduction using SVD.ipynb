```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import imsave
import numpy as np
from imageio import imread
import scipy
```

```python
image_path = 'pics/pic.jpg'
start= 0.5
steps = 0.1

img = imread(image_path, as_gray=True)
plt.imshow(img)
plt.suptitle('Original Image in gray scale')
plt.axis('off')
plt.show()

u, s, vh = np.linalg.svd(img, full_matrices=False)
threshold = start
while (threshold < 1.0):
    tmp = len(s)
    new_s  = s[:]
    for i in range (tmp):
        j = tmp - i - 1
        if (i >= tmp * threshold):
            break
        new_s[j] = 0.0
    denosed = np.array(np.dot(u * new_s, vh), dtype=int)
    plt.imshow(denosed) 
    plt.axis('off')
    plt.suptitle(f'threshold: {str(round(threshold,2))}')
    plt.show()
    imsave(image_path.split('.')[0] +'_'+ str(round(threshold,2))+'_denoised.jpg', denosed)
    threshold += steps
```
